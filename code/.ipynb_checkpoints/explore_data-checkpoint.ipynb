```python
%matplotlib inline
import numpy as np
import DFWnet
from matplotlib import pyplot as plt
```

```python
DFW = DFWnet.CommonData()
IPW = DFW.IPWvals
print IPW[159,...]
```

```python
temp_IPW = IPW[159,...]
print temp_IPW.shape
```

```python
temp_IPW[temp_IPW[:,0] == 'txno']
```

```python
no_data = ['okar']

IPWvals = temp_IPW

for temp in no_data:
    IPWvals = IPWvals[np.logical_not(IPWvals[:,0] == temp)]

print IPWvals[IPWvals[:,0] == 'txno',1:]
bad_sites = ['txno','txgl','txhm','txdc']
plt.figure()
ctr = 1
plt.subplots(2,2,figsize=(16,8))
for bs in bad_sites:
    plt.subplot(2,2,ctr)
    #print IPWvals[IPWvals[:,0] == bs,1:].reshape(-1,).shape
    raw_ipw = IPWvals[IPWvals[:,0] == bs,1:].astype('float').reshape(-1,)
    N = 12
    moving_average = np.convolve(raw_ipw,np.ones(N,)/N,mode='same')
    print moving_average.shape
    plt.plot(np.arange(48),raw_ipw,label=bs)
    plt.plot(moving_average)
    ctr+=1
```
